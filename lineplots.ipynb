# Plotting basics

## line plots

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

### toy data

```python
ds = xr.tutorial.load_dataset("air_temperature")
data_sample = ds["air"].sel(lon=270, lat=50)
```

```python
data_sample.plot()
```

### Smoothing by time averaging

```python
smooth = data_sample.resample(time="1MS").mean(dim="time")
```

```python
#plt.style.use("ggplot")
matplotlib.rcParams.update({"font.size": 18})
plt.figure(figsize=[6,4])
ax = plt.axes()
smooth.plot(ax=ax, color="black", linewidth=2)
ax.grid(color="black")
ax.set_title("Two seasonal cycles")
ax.set_xlabel("")
ax.set_ylabel(r"Air temperature [$^{\circ}K$]")
ax.set_yticks([250,260,270,280,290,300])
ax.set_yticklabels([250,260,270,280,290,300],
                   rotation=45, color="black")
ax.set_xticks([np.datetime64("2013-01-01"),
               np.datetime64("2014-01-01"),
               np.datetime64("2015-01-01")])
ax.set_xticklabels(["Jan\n2013", "Jan\n2014", "Jan\n2015"], rotation=50)
ax.set_xlim([np.datetime64("2013-01-01"), np.datetime64("2015-01-01")])
plt.show()
```

### multiple lines on one plot

```python
data_sample2 = ds["air"].sel(lon=290, lat=40)
smooth2 = data_sample2.resample(time="1MS").mean(dim="time")
```

```python
plt.style.use("default")
matplotlib.rcParams.update({"font.size": 18})
plt.figure(figsize=[6,4])
ax = plt.axes()
smooth.plot(ax=ax, linewidth=2, label="location 1")
smooth2.plot(ax=ax, linewidth=2, label="location 2")
ax.text(np.datetime64("2013-11-15"),295, "(290,40)", color="tab:orange")
ax.text(np.datetime64("2013-04-15"),260, "(270,50)", color="tab:blue")
ax.set_title("Two seasonal cycles")
ax.set_xlabel("")
ax.set_ylabel(r"Air temperature [$^{\circ}K$]")
ax.set_yticks([250,260,270,280,290,300])
ax.set_yticklabels([250,260,270,280,290,300],
                   rotation=45,)
ax.set_xticks([np.datetime64("2013-01-01"),
               np.datetime64("2014-01-01"),
               np.datetime64("2015-01-01")])
ax.set_xticklabels(["2013", "2014", "2015"], rotation=50)
ax.set_xlim([np.datetime64("2013-01-01"), np.datetime64("2015-01-01")])
ax.legend(bbox_to_anchor=[1.0,0.5], fontsize=14, ncols=2)
plt.show()
```

```python
plt.style.use("default")
matplotlib.rcParams.update({"font.size": 18})
plt.figure(figsize=[6,4])
ax = plt.axes()
ax2=ax.twinx()
smooth.plot(ax=ax, linewidth=2, label="location 1", color="tab:blue")
smooth2.plot(ax=ax2, linewidth=2, label="location 2", color="tab:orange")
#ax.grid(color="black")
ax.set_title("Two seasonal cycles")
ax2.set_title("")
ax.set_xlabel("")
ax.set_ylabel(r"Air temperature location 1 [$^{\circ}K$]", color="tab:blue")
ax2.set_ylabel(r"Air temperature location 2 [$^{\circ}K$]", color="tab:orange")
ax.set_yticks([250,260,270,280,290,300])
ax.set_yticklabels([250,260,270,280,290,300],
                   rotation=45, color="tab:blue")

ax2.set_yticks([250,260,270,280,290,300])
ax2.set_yticklabels([250,260,270,280,290,300],
                    rotation=-45, color="tab:orange")


ax.set_xticks([np.datetime64("2013-01-01"),
               np.datetime64("2014-01-01"),
               np.datetime64("2015-01-01")])
ax.set_xticklabels(["2013", "2014", "2015"], rotation=50)




ax.set_xlim([np.datetime64("2013-01-01"), np.datetime64("2015-01-01")])
plt.show()
```

### subplots, scatter and fit

```python
def plot_fit(ax, pols, xrange=[240, 320], kwargs=None):
    x = np.arange(xrange[0], xrange[1])
    ax.plot(x, pols[1] + x*pols[0], **kwargs)
    return None

def set_props(ax):
    ax.set_xlim(240,310)
    ax.set_ylim(240,310)
    ax.set_xticks(np.arange(250,310+20,20))
    ax.set_yticks(np.arange(250,310+20,20))
    ax.grid()
    ax.set_xlabel("location 1", style="italic")
    ax.set_ylabel("location 2", style="italic")
    return None
    
f, axs = plt.subplots(ncols=2, figsize=[12,4]) 

axs[0].scatter(data_sample, data_sample2, color="grey")
pols = np.polyfit(data_sample, data_sample2, 1)
plot_fit(axs[0], pols, kwargs=dict(color="red", linewidth=2))
R = np.corrcoef(data_sample, data_sample2).min()
axs[0].text(250,250, f"R = {R:0.2f}", color="red",
           bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
set_props(axs[0])

axs[1].scatter(smooth, smooth2, color="grey")
pols = np.polyfit(smooth, smooth2, 1)
plot_fit(axs[1], pols, kwargs=dict(color="red", linewidth=2))
R = np.corrcoef(smooth, smooth2).min()
axs[1].text(250,250, f"R = {R:0.2f}", color="red",
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
set_props(axs[1])

plt.subplots_adjust(wspace=0.3)
```
